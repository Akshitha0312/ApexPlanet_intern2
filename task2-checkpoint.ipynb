```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("C:\\Users\\prash\\OneDrive\\docs\\Documents\\Desktop\\ApexPlanet_DataAnalytics_Dataset.xlsx")
```

```python
df.head()

df.info()

df.describe()
```

```python
print("Total Sales:", df['Total_Sales'].sum())
print("Average Sales:", df['Total_Sales'].mean())
print("Total Quantity:", df['Quantity'].sum())
```

```python
# Sales distribution

plt.figure(figsize=(8,5))
sns.histplot(df['Total_Sales'], bins=20)
plt.title("Distribution of Sales")
plt.show()
```

```python
# category count

df['Category'].value_counts().plot(kind='bar')
plt.title("Products by Category")
plt.show()
```

```python
# Gender distribution

df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
# top 5 products by sales

top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)

print(top_products)
```

```python
# Top cities by revenue

top_cities = df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False)

print(top_cities)
```

```python
# Category wise sales

category_sales = df.groupby('Category')['Total_Sales'].sum()

print(category_sales)
```

```python
### Category vs Sales

plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Total_Sales', data=df)
plt.show()
```

```python
### Age vs Sales

plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Total_Sales', data=df)
plt.show()
```

```python
### Correlation Analysis

corr = df[['Age','Quantity','Unit_Price','Total_Sales']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Matrix")
plt.show()
```

```python
total_revenue = df['Total_Sales'].sum()
total_orders = df['Order_ID'].nunique()
average_order_value = df['Total_Sales'].mean()
top_category = df.groupby('Category')['Total_Sales'].sum().idxmax()
top_city = df.groupby('City')['Total_Sales'].sum().idxmax()

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)
print("Average Order Value:", average_order_value)
print("Top Category:", top_category)
print("Top City:", top_city)
```
